```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.Department = le.fit_transform(dfle.Department)
dfle.salary = le.fit_transform(dfle.salary)
dfle
```

```python
X = dfle[['department','area']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfle[['salary','Department','satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident',]],dfle.left,train_size=0.8)
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
